```python
k = 500
grid = range(0, 500)
arr = [10, 100, 500]
```

```python
# Положим sigma = 7
sigma = 7
```

```python
# для графиков
j = 1
```

```python
plt.figure(figsize=(16, 12))

for n in arr:
    estimate_first, estimate_second = [], []
    for i in grid:
        distribution = np.random.normal(0, sigma, size=n)
        estimate_first.append((np.square(distribution).mean() - (distribution.mean())**2))
        estimate_second.append((k / (k - 1)) * (np.square(distribution).mean() - (distribution.mean())**2))

    first_mean = np.array(estimate_first).mean()
    second_mean = np.array(estimate_second).mean()

    plt.subplot(3, 1, j)

    plt.scatter(estimate_first, np.zeros(k),
                alpha=0.1, s=100, color='blue', label=r'$\sigma^2$')
    plt.scatter(first_mean, 0, marker='*', s=200,
                color='w', edgecolors='black')

    plt.scatter(estimate_second, np.zeros(k) + 1,
                alpha=0.1, s=100, color='green', label=r'$\frac{n}{n-1}\sigma^2$')
    plt.scatter(second_mean, 1, marker='*', s=200,
                color='w', edgecolors='black')

    plt.title(f'$n = {n}$')
    plt.vlines(sigma**2, 0, 1, color='r', lw=4)
    plt.yticks([])
    plt.legend(loc='upper left')
    plt.xlim(sigma**2 - 4, sigma**2 + 4)

    j += 1
plt.suptitle(f'Оценки для $sigma^2$ = {sigma**2}')
```

**2.** Изучим поведение среднего оценок из первого пункта при росте размера выборки. Постройте график зависимости $\widetilde{\theta}$ от $n$ для двух типов оценок. Для вычисления зависимости нужно один раз сгенерировать выборки из пункта 1 достаточно большого размера (не более 500) и посчитать оценки по подвыборкам, используя <a href="https://miptstats.github.io/courses/python/05_numpy.html#2.1-Математические-операции">функции</a> из `numpy`. Использовать циклы, а так же функции, разворачивающиеся в цикл (например, `np.vectorize`), запрещено.

**Решение:**

```python
n = k = 500
sigma = 3
```

```python
distribution = np.random.normal(0, sigma, size=(n, k))

distributions = np.array(n * [distribution])
distributions_square = np.array(n * [np.square(distribution)])
tri = np.array(n * [np.tri(n)])
```

```python
means = np.sum((distributions * tri.T).T, axis=1) / np.flip(np.cumsum(np.ones(n)))
means_square = np.sum((distributions_square * tri.T).T, axis=1) / np.flip(np.cumsum(np.ones(n)))
```

```python
estimate_first = np.flip((means_square - np.square(means)).mean(axis=0))
estimate_second = np.flip(((means_square - np.square(means)).mean(axis=0))[:(n - 1)] * np.flip(np.cumsum(np.ones(n)))[:(n - 1)] / \
                           (np.flip(np.cumsum(np.ones(n))) - 1)[:(n - 1)])
```

```python
first_grid = range(1, n + 1)
second_grid = range(2, n + 1)
```

```python
plt.figure(figsize=(15, 5))

plt.scatter(first_grid, estimate_first, color='cyan', label=r'$\sigma^2$', alpha=0.1)
plt.scatter(second_grid, estimate_second, color='orange', label=r'$\frac{n}{n-1}\sigma^2$', alpha=0.1)
plt.xlabel('$n$')
plt.ylabel('Средняя оценка')
plt.legend(loc='upper right', shadow=True)
plt.title(rf'$\sigma^2={3**2}$')
plt.ylim(sigma**2 - 0.5, sigma**2 + 0.5)
plt.xlim(0, n + 1)

plt.hlines(sigma**2, 0, 500, color='red', lw=2)
```
